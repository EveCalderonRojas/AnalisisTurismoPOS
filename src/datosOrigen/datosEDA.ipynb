### ASIGNAR LAS CATEGORÍAS A CADA LUGAR

```python
# Diccionario de categorías por lugar
categorias = {
    'Playa Manuel Antonio'       : 'parque',
    'El Yugo de Mi Tata'         : 'restaurante',
    'Playa Chiquita'             : 'parque',
    'Restaurante Mi Finca'       : 'restaurante',
    'Monteverde Country Lodge'   : 'alojamiento',
    'Catarata Río Fortuna'       : 'parque',
    'Hotel Los Lagos Spa & Resort': 'alojamiento',
    'Restaurante Fortuneño'      : 'restaurante',
    'Corcovado National Park'    : 'parque',
}
```

```python
# CORPUS CON SUS RESPECTIVAS CATEGORÍAS DE LUGAR
corpus_clean = pd.DataFrame({
    'lugar'        : corpus_raw['title'].str.strip(),
    'texto'        : corpus_raw['text'],
    'calificacion' : corpus_raw['stars'],
    'categoria'    : corpus_raw['title'].map(categorias),
    'fuente'       : 'google_maps'
})


corpus_clean.head(15)
```

### TRADUCCIÓN DE COMENTARIOS AL ESPAÑOL

```python
from langdetect import detect
from langdetect.lang_detect_exception import LangDetectException
from deep_translator import GoogleTranslator
```

```python
# IDENTIFICA EL IDIOMA DEL COMENTARIO
# SI NO LO PUEDE DETECTAR LO CATEGORIZA COMO 'UNKNOWN'
def idioma_confiable(texto):
    if pd.isna(texto) or len(texto.split()) < 3:
        return "unknown"
    try:
        return detect(texto)
    except LangDetectException:
        return "unknown"
```

```python
# AGREGA EL IDIOMA DETECTADO EN UNA NUEVA COLUMNA Y HACE UN CONTEO DE LOS IDIOMAS PRESENTES

corpus_clean["idioma_comentario"] = corpus_clean["texto"].apply(idioma_confiable)
corpus_clean["idioma_comentario"].value_counts()
```

```python
# FUNCIÓN PARA TRADUCIR EL TEXTO ORIGINAL AL ESPAÑOL
translator = GoogleTranslator(source="auto", target="es")

def traducir_texto(texto, idioma):
    if pd.isna(texto) or texto.strip() == "":
        return texto
    if idioma in ["es", "unknown"]:
        return texto
    try:
        return translator.translate(texto)
    except:
        return texto
```

```python
# AGREGA EL COMENTARIO YA TRADUCIDO AL CORPUS EN UNA NUEVA COLUMNA
# NO SE MODIFICA EL COMENTARIO ORIGINAL

corpus_clean["comentarios_espanol"] = corpus_clean.apply(
    lambda row: traducir_texto(row["texto"], row["idioma_comentario"]), axis=1
)
```

```python
# SEPARAMOS LOS EMOJIS DEL COMENTARIO

import re

def extraer_emojis(texto):
    if pd.isna(texto):
        return ""
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticonos
        u"\U0001F300-\U0001F5FF"  # símbolos y pictogramas
        u"\U0001F680-\U0001F6FF"  # transporte y mapas
        u"\U0001F1E0-\U0001F1FF"  # banderas
        u"\U00002700-\U000027BF"  # dingbats
        u"\U0001F900-\U0001F9FF"  # símbolos adicionales
        "]+", flags=re.UNICODE)
    return " ".join(emoji_pattern.findall(texto))

def quitar_emojis(texto):
    if pd.isna(texto):
        return texto
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF"
        u"\U00002700-\U000027BF"
        u"\U0001F900-\U0001F9FF"
        "]+", flags=re.UNICODE)
    return emoji_pattern.sub("", texto).strip()


# CADA EMOJI SE COLOCA EN UNA NUEVA COLUMNA

corpus_clean["emojis"] = corpus_clean["comentarios_espanol"].apply(extraer_emojis)
corpus_clean["comentarios_espanol"] = corpus_clean["comentarios_espanol"].apply(quitar_emojis)
```
