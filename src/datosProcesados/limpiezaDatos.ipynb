# Limpieza principal del corpus


### IMPORTACIÓN DE LIBRERÍAS Y CORPUS

```python
import pandas as pd
```

```python
corpus_raw = pd.read_csv("../../data/processed/dataset_Google-Maps-Reviews-Scraper.csv_clean.csv")

corpus_raw.head(20)
```

### LIMPIEZA DEL CORPUS


```python
# IMPORTAMOS LAS FUNCIONES

from limpieza import (
    filtrar_unknown,
    aplicar_limpieza
)
```

```python
corpus = filtrar_unknown(corpus_raw)

print(f'Comentarios totales   : {len(corpus_raw)}')
print(f'Comentarios unknown   : {len(corpus_raw) - len(corpus)}')
print(f'Comentarios a analizar: {len(corpus)}')
```

```python
corpus.head(20)
```

```python
# CORPUS SIN LOS COMENTARIOS EN CATEGORÍA 'UNKNOWN'

corpus_comentarios = pd.DataFrame(corpus)
corpus_comentarios.to_csv('../../data/processed/corpus_Google-Maps-Reviews.csv', index=False)

print(f'Archivo guardado: corpus_Google-Maps-Reviews.csv')
```

```python
extras = set()

todas_las_palabras, stop_words = aplicar_limpieza(corpus, extras_stopwords=extras)

print(f'Total palabras procesadas: {len(todas_las_palabras)}')
print(f'Vocabulario único        : {len(set(todas_las_palabras))}')
```

```python
# CSV CON LAS PALABRAS MÁS REPETIDAS DEL CORPUS

comentarios_procesados = pd.DataFrame(todas_las_palabras, columns=['palabras'])
comentarios_procesados.to_csv('../../data/processed/corpus_palabras.csv', index=False)

print(f'Archivo guardado: corpus_palabras.csv')
print(f'Total filas: {len(comentarios_procesados)}')
```
