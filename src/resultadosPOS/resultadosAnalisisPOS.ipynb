# Resultados del Análisis Universal POS / Penn Treebank


### IMPORTACIÓN DE LIBRERÍAS Y LLAMADO A LAS FUNCIONES

```python
import pandas as pd

import sys
sys.path.append('../analisisPOS')

from posTreebank import (
    pos_treebank,
    resumen_treebank)

from posUniversal import (
    pos_universal,
    resumen_universal
)
```

```python
corpus = pd.read_csv('../../data/processed/corpus_Google-Maps-Reviews.csv')

corpus.head(10)
```

## PENN TREEBANK CON NLTK

```python
# ANÁLISIS DE PENN TREEBANK

analisis_treebank = pos_treebank(corpus)

analisis_treebank.head(10)
```

## UNIVERSAL POS CON SPACY

```python
# ANÁLISIS DE UNIVERSAL POS
analisis_universal = pos_universal(corpus)

analisis_universal.head(5)
```

```python
# GUARDADO DE LOS RESULTADOS DEL ANÁLISIS EN UN CORPUS NUEVO

corpus_analisis = corpus.copy() # COPIA DEL ORIGINAL

# INCLUSIÓN DE LAS COLUMNAS CORRESPONDIENTES CON LOS RESULTADOS
corpus_analisis['penntreebank']  = analisis_treebank['PennTreebank']
corpus_analisis['universalpos'] = analisis_universal['UniversalPOS']


# GUARDADO DEL CORPUS EN LA RUTA DETERMINADA
ruta_processed = "../../data/processed/dataset_Google-Maps-Reviews-Scraper-postagging.csv"

corpus_analisis.to_csv(ruta_processed, index=False)
print("Archivo guardado en data/processed")
```
